```python
import matplotlib.pyplot as plt

import numpy as np
```

```python
# dummy data
labels = ["0.01", "0.1", "1", "10"]
bitsize = 10000
upswitch = np.array((1, 7, 53, 342)) / bitsize
downswitch = np.array((1, 5, 46, 310)) / bitsize
width = 0.6
```

```python
plt.style.use("ggplot")

fig = plt.figure(figsize=(8,4))
ax = fig.subplots()

ax.bar(labels, upswitch, width, label='bitflip 0 -> 1')
ax.bar(labels, downswitch, width, bottom=upswitch, label='bitflip 1 -> 0')

#ax.set_xscale('log')

plt.title("Bit Flips of SSD Drives")
plt.xlabel("radiation dose / Gy")
plt.ylabel("flipped bits / %")
plt.legend(loc="upper left")

plt.savefig("result-simulation-1.pdf")
plt.show()
```
